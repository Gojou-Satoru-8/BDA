```python
# Q1. Read Total profit of all months and show it using a line plot.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/sample_data/company_sales_data.csv")

total_profit = df["total_profit"]

plt.plot(df["month_number"], total_profit)
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.show()
```

```python
# Q2. Read all product sales data and show it using a multiline plot

# plt.plot(df["facecream"], label="facecream")
# plt.plot(df["facewash"], label="facewash")
# plt.plot(df["toothpaste"], label="toothpaste")
# plt.plot(df["bathingsoap"], label="bathingsoap")
# plt.plot(df["shampoo"], label="shampoo")
# plt.plot(df["moisturizer"], label="moisterizer")

columns = list(filter(lambda x: x not in ["month_number", "total_units", "total_profit"], df.keys()))
print(columns)
for column in columns:
  plt.plot(df["month_number"], df[column], label=column)

plt.legend(loc="upper left")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.show()
```
